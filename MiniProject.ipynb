```
from skimage.io import imread
from skimage.transform import resize
```

```
target=[]#This is the list of output
flat_data=[]#Image pixel in the flatten style
images=[]
path='/content/drive/My Drive/SmartknowerML/images/images/'
```

```
for  dirpath,dirnames,filenames in os.walk(path):
  print(filenames)
```

```
for  dirpath,dirnames,filenames in os.walk(path):
  print(dirnames)
```

```
for  dirpath,dirnames,filenames in os.walk(path):
  print(dirpath)
```

```
import numpy as np
for  dirpath,dirnames,filenames in os.walk(path):
   for filename in filenames:
    category=dirpath.split('/')[-1]
    target.append(category)
    img=imread(f'{path}{category}/{filename}')
    img_resize=resize(img,(100,100,3))
    flat_data.append(img_resize.flatten())
    images.append(img_resize)

flat_data=np.array(flat_data)
target=np.array(target)
images=np.array(images)
print(flat_data)
print(target)

print(flat_data.shape)
print(target.shape)
print(images.shape)
```

```
import pandas as pd
df=pd.DataFrame(flat_data)
df['Target']=target
df.head()
```

```
df.shape
```

```
x=flat_data
y=target
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)

(x_train).shape
```

```
x_test.shape
```

```
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```
y_test#For verification.
```

```
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_pred,y_test)
```

```
confusion_matrix(y_pred,y_test)
```
